# **Theory Questions**

1. What is hypothesis testing in statistics?
   - Hypothesis Testing is a statistical method used to make decisions or inferences about population parameters based on sample data.
   - a. Null Hypothesis (H₀): A default assumption that there is no effect or no difference.
       - Example: "There is no difference in average exam scores between two classes."
  - Alternative Hypothesis (H₁ or Ha): The statement we want to test or prove.
      - Example: "There is a difference in average exam scores between two classes."
  - b. Test Statistic: A value calculated from sample data used to determine whether to reject the null hypothesis.
  - c. P-value: The probability of getting the observed result if the null hypothesis is true.
       If p-value < significance level (e.g., 0.05) - reject H₀
       If p-value ≥ significance level - fail to reject H₀
  - Significance Level (α): The threshold (commonly 0.05) for rejecting the null hypothesis.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
   - In statistics, hypotheses are assumptions we make about a population parameter, which we test using sample data.
   - Null Hypothesis (H₀): It is the default assumption that there is no effect, no difference, or no relationship.
          - It represents a status quo or a claim to be tested.
          - We try to disprove or reject it based on evidence from sample data.
          - Example: Suppose a company claims its battery lasts 10 hours - H₀: Battery life = 10 hours
   - Alternative Hypothesis (H₁ or Ha): It is what we want to prove - that there is an effect, difference, or relationship.
         - It's accepted when there's strong evidence against the null hypothesis.
         - Example: We suspect the battery lasts less than 10 hours - H₁: Battery life < 10 hours

3. What is the significance level in hypothesis testing, and why is it important?
   - The significance level (denoted by α) is a threshold used in hypothesis testing to determine whether to reject the null hypothesis. It sets the standard for evidence needed to reject the null hypothesis. Choosing it carefully ensures valid conclusions in statistical testing.
   - Significance level (α) is the probability of rejecting the null hypothesis when it is actually true (Type I error).
   - It represents the maximum acceptable risk of making a wrong decision.
         - 0.05 - 5% chance of error (most commonly used)
         - 0.01 - 1% chance of error (for stricter tests)
         - 0.10 - 10% chance of error (for more lenient decisions)
   -  Decision Boundary: Helps determine whether the test result is statistically significant.
   - Controls Type I Error: Keeps the risk of wrongly rejecting a true null hypothesis under control.
   - Affects P-value Interpretation:
              - If P-value < α, reject H₀
              - If P-value ≥ α, fail to reject H₀

4. What does a P-value represent in hypothesis testing?
   - The P-value is the probability of observing the test results, or something more extreme, assuming that the null hypothesis is true.
   - Smaller P-value = stronger evidence against H₀.
   - t measures the strength of evidence against the null hypothesis (H₀).
   - A small P-value indicates that the observed data is unlikely under H₀ - strong evidence against H₀.
   - A large P-value suggests the data is likely under H₀ - weak evidence against H₀.

5. How do you interpret the P-value in hypothesis testing?
   - The P-value indicates the probability of getting the observed results, or more extreme, if the null hypothesis is true.
        - Low P-value (≤ 0.05): Strong evidence against the null hypothesis - reject H₀.
        - High P-value (> 0.05): Weak evidence against the null - fail to reject H₀.

6. What are Type 1 and Type 2 errors in hypothesis testing?
   - Type I Error (False Positive): Rejecting the null hypothesis (H₀) when it is actually true.
   - Example: Concluding a drug works when it doesn't.
   - Controlled by significance level (α), e.g., 0.05.
   - Type II Error (False Negative): Failing to reject the null hypothesis when it is actually false.
   - Example: Concluding a drug doesn't work when it actually does.
   - Related to power of the test (1 - β).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
   - One-Tailed Test: Tests for the effect in one specific direction (greater than or less than).
      - Example: H₁: μ > μ₀ or H₁: μ < μ₀
      - Used when we have a directional hypothesis.
   - Two-Tailed Test: Tests for the effect in both directions (not equal to).
      - Example: H₁: μ ≠ μ₀
      - Used when we want to detect any difference, regardless of direction.
      - One-tailed test has more power to detect an effect in one direction.
      - Two-tailed test is more conservative and widely used when no direction is specified.

8. What is the Z-test, and when is it used in hypothesis testing?
   - Z-test is a statistical test used to determine whether there is a significant difference between sample and population means (or between two sample means) when the population variance is known and the sample size is large (typically n > 30).
   - It is used:
     - When the population standard deviation (σ) is known.
     - The sample size is large (n ≥ 30) or the population is normally distributed.
     - To test mean differences (e.g., sample mean vs. population mean).

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
   - Z= X - μ /σ/ root of n
   - Where:

          X = Sample mean
          μ = Population mean
          σ = Population standard deviation
          n = Sample size
  - The Z-score shows how many standard errors our sample mean is from the population mean.
  - It helps determine whether to reject or fail to reject the null hypothesis.- A high absolute Z-score suggests the sample is significantly different from the population mean.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
    - The T-distribution (or Student's t-distribution) is a probability distribution used in statistics when estimating population parameters with a small sample size and/or unknown population standard deviation.
    - When to Use the T-Distribution Instead of the Normal Distribution: Use the T-distribution when:
    - The sample size is small (typically n < 30).
    - The population standard deviation (σ) is unknown.
    - The population is approximately normally distributed.
    - The T-distribution is wider and has heavier tails, accounting for more variability in small samples.
    - As sample size increases, the T-distribution approaches the normal distribution.

11. What is the difference between a Z-test and a T-test?
    - Use Z-test when:
        - Population variance is known
        - Sample size is large (n > 30)
    - Use T-test when:
        - Population variance is unknown
        - Sample size is small (n ≤ 30)
    - Z-distribution is normal, T-distribution is wider (accounts for more variability)

12. What is the T-test, and how is it used in hypothesis testing?
    - A T-test is a statistical test used to compare the mean(s) of one or two groups to determine if the difference is significant.
    - a. Set hypotheses:
            - H₀ (Null): No difference in means
            - H₁ (Alt): There is a difference
   - b. Choose test type:
            - One-sample T-test: Compare sample mean vs population mean
            - Two-sample T-test: Compare means of two independent groups
            - Paired T-test: Compare means of the same group before & after
    - c. Calculate T-statistic:
            - Measures how far sample mean is from expected mean, in units of standard error
    - Check p-value:
            - If p < α (e.g., 0.05) - Reject H₀
            - If p ≥ α - Fail to reject H₀

13. What is the relationship between Z-test and T-test in hypothesis testing?
    - Both are used to test hypotheses about population means.
    - Both compare the sample statistic vs population parameter.
    - T-test becomes Z-test as sample size increases (n - ∞).
    - Z-test assumes known population variance (σ),
    - T-test is used when σ is unknown.
    - Both give a test statistic, compared against critical values or p-values to make decisions.

14. What is a confidence interval, and how is it used to interpret statistical results?
    - A confidence interval gives a range of values within which the true population parameter (like mean or proportion) is expected to fall, with a certain level of confidence (e.g., 95%).
    - Interpretation: A 95% CI means: “We are 95% confident that the true value lies within this range.”
    - In Hypothesis Testing: If the CI does not include the null hypothesis value (e.g., mean = 0), we reject H₀. If it includes the null value, we fail to reject H₀.
    - Mean score = 70, 95% CI = [66, 74] - We're 95% confident that the true mean is between 66 and 74.

15. What is the margin of error, and how does it affect the confidence interval?
    - The margin of error is the maximum expected difference between the sample estimate and the true population parameter. It defines the radius (half-width) around the sample statistic in a confidence interval.
    - Margin of error controls the width of the confidence interval and reflects uncertainty in the estimate.
    - CI=Sample Statistic ± Margin of Error
    - Confidence Interval = Sample Estimate ± Margin of Error
        - Larger ME = Wider CI - Less precise estimate
        - Smaller ME = Narrower CI - More precise estimate

16. How is Bayes' Theorem used in statistics, and what is its significance?
    - Bayes' Theorem lets us update probabilities rationally as we get new info, making it crucial for real-world decision-making under uncertainty.
    - Helps revise beliefs when new data arrives (Bayesian updating)
    - Widely used in: Medical diagnosis (probability of disease given test results)
    - Machine learning (classification, spam filtering)
    - Risk assessment and decision making
    - Bridges prior knowledge and current evidence

17. What is the Chi-square distribution, and when is it used?
    - A right-skewed probability distribution used mainly for testing relationships between categorical variables.
    - Chi-square distribution helps test associations and fits for categorical data by comparing observed vs expected counts.
    - It's the distribution of a sum of squared standard normal variables.
    - Chi-square test of independence: To check if two categorical variables are related (e.g., gender vs. voting preference).
    - Goodness-of-fit test: To see if observed data fits an expected distribution.
    - Test for variance: To test if a population variance equals a specified value (in normal populations).

18. What is the Chi-square goodness of fit test, and how is it applied?
    - A statistical test used to determine whether observed categorical data matches an expected distribution.
    - Set hypotheses: H₀: Data fits the expected distribution, H₁: Data does not fit the expected distribution
    - Calculate expected counts for each category based on the hypothesized distribution.
    - Compute test statistic
    - Compare χ2 value with critical value from Chi-square distribution (degrees of freedom = number of categories - 1).
    - Decision: If χ2 > critical value - Reject H₀ (poor fit), Else - Fail to reject H₀ (good fit)

19. What is the F-distribution, and when is it used in hypothesis testing?
    - F-distribution helps test if variances or model effects are statistically significant by comparing variability across groups.
    - A right-skewed distribution used to compare two variances or test overall model significance (like in ANOVA or regression).
    - ANOVA (Analysis of Variance): To check if group means differ significantly
           - F = variance between groups / variance within groups
    - Regression Analysis: To test if the overall model is significant
        - H₀: All regression coefficients = 0
    - Compare Two Variances: Test if two populations have equal variances

20. What is an ANOVA test, and what are its assumptions?
    - A statistical method used to compare means of 3 or more groups to see if at least one group mean differs significantly.
    - Tests the overall difference among group means using the F-statistic.
    - Assumptions of ANOVA:
       - Independence: Observations are independent within and across groups.
       - Normality: Data in each group is approximately normally distributed.
       - Homogeneity of variances: All groups have equal variances (also called homoscedasticity).

21. What are the different types of ANOVA tests?
    - One-Way ANOVA: Tests mean differences across one categorical factor with 3+ groups.
      - Example: Comparing test scores across 3 different teaching methods
    - Two-Way ANOVA: Tests effects of two categorical factors and their interaction on the mean.
      - Example: Effect of teaching method and gender on test scores.
    - Repeated Measures ANOVA: Used when the same subjects are measured multiple times under different conditions.
       - Example: Measuring blood pressure of patients before, during, and after treatment.
    - Multivariate ANOVA (MANOVA): Tests for differences in multiple dependent variables across groups.
       - Example: Comparing diet effects on weight and cholesterol simultaneously.

22. What is the F-test, and how does it relate to hypothesis testing?
    - A statistical test used to compare variances between two or more groups or to test the overall significance in models like ANOVA (Analysis of Variance).
    - F-test checks if group variances or means differ significantly by comparing variance ratios under the null hypothesis.
    - Null hypothesis (H₀): All group variances are equal (or no difference in group means)
    - Alternative hypothesis (H₁): At least one variance (or mean) differs
    - The F-statistic is the ratio of variances (between-group variance / within-group variance).
    - If F is large and p-value < significance level (e.g., 0.05), reject H₀ - evidence of difference in variances or means.
    - Use cases: Comparing variances of two samples, Testing overall model significance in regression or ANOVA

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        return 0
    return p if x == 1 else 1 - p

print("P(X=0):", bernoulli_pdf(0, 0.7))
print("P(X=1):", bernoulli_pdf(1, 0.7))
```

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm

# Population parameters
population_mean = 50
population_std = 10

# Sample data
sample = [52, 49, 51, 48, 50, 53, 47, 52, 49, 51]
sample_size = len(sample)
sample_mean = np.mean(sample)

# Calculate Z-statistic
z = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
print(f"Z-statistic: {z:.2f}")

# Significance level (for 95% confidence, alpha = 0.05)
alpha = 0.05
z_critical = norm.ppf(1 - alpha / 2)

# Hypothesis testing
print(f"Z-critical (±): {z_critical:.2f}")
if abs(z) > z_critical:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy import stats

# Sample data
data = [52, 49, 51, 48, 50, 53, 47, 52, 49, 51]
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation
n = len(data)

# Confidence level
confidence = 0.95
alpha = 1 - confidence

# Calculate margin of error using t-distribution
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# We can be 95% confident that the true population mean lies between the lower and upper bounds of the interval.
```

```python
# 24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(size=1000, mean=0, std=1):
    # Generate normal distribution data
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness
    skewness = skew(data)

    # Calculate kurtosis (Fisher's definition: normal dist = 0)
    kurt = kurtosis(data)

    return skewness, kurt

# Example usage
skewness, kurtosis_value = calculate_skewness_kurtosis()
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurtosis_value:.4f}")
```

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Calculate Z-statistic
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Interpret the result at 5% significance level
    if p_value < 0.05:
        result = "Reject the null hypothesis (significant difference)."
    else:
        result = "Fail to reject the null hypothesis (no significant difference)."

    return z_stat, p_value, result

# Example
np.random.seed(0)
sample = np.random.normal(loc=52, scale=10, size=30)  # sample from population with true mean 52
z, p, conclusion = z_test(sample_data=sample, population_mean=50, population_std=10)

print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p:.4f}")
print("Conclusion:", conclusion)
```

```python
# 2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.
import numpy as np
from scipy.stats import norm

# Generate sample data (mean=51, std=10, size=40)
data = np.random.normal(51, 10, 40)

# Known population mean and std
pop_mean = 50
pop_std = 10

# Calculate sample mean and size
sample_mean = np.mean(data)
n = len(data)

# Z-test
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p = 2 * (1 - norm.cdf(abs(z)))

# Result
print("Z-score:", round(z, 2))
print("P-value:", round(p, 4))

if p < 0.05:
    print("Significant: Reject null hypothesis")
else:
    print("Not significant: Fail to reject null hypothesis")
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def simple_error_plot():
    # Assume: population std = 10, sample size = 30
    sigma = 10
    n = 30
    se = sigma / np.sqrt(n)

    # H0: mean = 50, H1: mean = 55
    mu0 = 50
    mu1 = 55
    alpha = 0.05

    # Critical value for one-tailed test (right)
    z_crit = norm.ppf(1 - alpha)
    x_crit = mu0 + z_crit * se

    # X values for plotting
    x = np.linspace(40, 65, 1000)
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    # Plot
    plt.plot(x, y0, label="Null (μ=50)")
    plt.plot(x, y1, label="Alternative (μ=55)")

    # Shade Type I Error
    plt.fill_between(x, y0, where=(x > x_crit), color='red', alpha=0.3, label="Type I Error (α)")

    # Shade Type II Error
    plt.fill_between(x, y1, where=(x < x_crit), color='blue', alpha=0.3, label="Type II Error (β)")

    # Line for critical value
    plt.axvline(x_crit, color='black', linestyle='--', label=f"Critical value = {x_crit:.2f}")

    plt.title("Type I and Type II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

simple_error_plot()
```

```python
# 8. Simulate data and perform both Z-test and T-test, then compare the results using Python.
import numpy as np
from scipy.stats import norm, t

# Simulate data
np.random.seed(42)
data = np.random.normal(loc=52, scale=10, size=40)  # Sample data

# Known population parameters
pop_mean = 50
pop_std = 10  # known population std dev

# Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

# Z-test calculation
z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value_z = 2 * (1 - norm.cdf(abs(z_stat)))  # two-tailed

# T-test calculation
t_stat = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))
p_value_t = 2 * (1 - t.cdf(abs(t_stat), df=n-1))  # two-tailed

# Print results
print(f"Sample mean: {sample_mean:.2f}")
print(f"Z-test statistic: {z_stat:.3f}, p-value: {p_value_z:.4f}")
print(f"T-test statistic: {t_stat:.3f}, p-value: {p_value_t:.4f}")

# Interpretation
alpha = 0.05
print("\nInterpretation:")
print(f"Z-test: {'Reject H0' if p_value_z < alpha else 'Fail to reject H0'}")
print(f"T-test: {'Reject H0' if p_value_t < alpha else 'Fail to reject H0'}")
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # standard error

    # t-critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    return lower, upper

# Example usage
sample_data = [12, 15, 14, 10, 13, 12, 16, 14]
ci_lower, ci_upper = confidence_interval(sample_data)

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    n = len(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # standard error
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)  # t critical value
    moe = t_crit * std_err
    return moe

# Example data
sample_data = [10, 12, 14, 13, 15, 11, 16, 14]

moe = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error at 95% confidence: {moe:.3f}")
```

```python
# 13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
import numpy as np

# Observed data (contingency table)
observed = np.array([[30, 10],
                     [20, 40]])

# Calculate row totals and column totals
row_totals = observed.sum(axis=1).reshape(-1, 1)  # sum for each row (as column vector)
col_totals = observed.sum(axis=0).reshape(1, -1)  # sum for each column (as row vector)
total = observed.sum()

# Calculate expected frequencies
expected = (row_totals @ col_totals) / total

print("Observed frequencies:")
print(observed)
print("\nExpected frequencies:")
print(expected)
```
